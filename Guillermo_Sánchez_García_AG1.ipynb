```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

# Encontrar los dos puntos más cercanos - Dimensión 1 (Divide y Vencerás)

```python
def cercanos_1D_divide_y_venceras(lista):
  if len(lista) < 2:
    return 'No hay un mínimo de dos puntos'

  # El algoritmo se va basará en ir diviendo la lista en mitades para luego compararlas
  # Primero ordenamos la lista

  lista_ordenada = sorted(lista)

  def resolver(sublista):
    # Iremos reduciendo la lista hasta tener grupos de 2 ( lista original tenga cardinal par)
    # o en grupos de 3 (por si la lista original es impar, alguna vez habrá tres)

    if len(sublista) <= 3:
      minimo = sublista[1] - sublista[0]
      ele1 = sublista[0]
      ele2 = sublista[1]
      for i in range(len(sublista)):
        for j in range(i+1,len(sublista)):
          if( minimo > abs(sublista[i] - sublista[j])):
            minimo = abs(sublista[i] - sublista[j])
            ele1 = sublista[i]
            ele2 = sublista[j]
      return ele1, ele2, minimo


    # Encontramos el punto medio
    mitad = len(sublista) // 2
    ele1_izq, ele2_izq, min_izq = resolver(sublista[:mitad])
    ele1_der, ele2_der, min_der = resolver(sublista[mitad:])

    # Elegimos el mínimo entre  la mitad derecha y la mitad izquierda
    if min_izq < min_der:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_izq, ele2_izq, min_izq
    else:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_der, ele2_der, min_der

    # Al separar en dos mitades,
    # el último número de la izquierda y el primero de la derecha son contiguos,
    # tendremos que ver si la distancia entre ellos es menor que la que hemos obtenido de comparar las dos mitades

    distancia = sublista[mitad] - sublista[mitad-1]

    if distancia < minimo_elegido:
      return sublista[mitad-1], sublista[mitad], distancia

    return ele1_elegido, ele2_elegido , minimo_elegido

  resultado =  resolver(lista_ordenada)

  return resultado[0], resultado[1]
```

```python
LISTA_1D = [random.randrange(1,10000) for x in range(1000)]
print(LISTA_1D)
print(cercanos_1D_divide_y_venceras(LISTA_1D))
```

Como hemos utilizado un algoritmo por divide y venceras, el código tiene complejidad $O(nlog(n))$, donde no podemos reducirla.

# Encontrar los dos puntos más cercanos - Dimensión 2 (Divide y Vencerás)

```python
def cercanos_2D_divide_y_venceras(lista):

  if len(lista) < 2:
    return 'No hay un mínimo de dos puntos'

  # Primero ordenamos la lista sobre el eje x
  lista_ordenada = sorted(lista, key=lambda p: p[0])

  def distancia_2D(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

  def resolver(sublista):
    # Iremos reduciendo la lista hasta tener grupos de 2 o 3
    if len(sublista) <= 3:
      # ERROR CORREGIDO: Usamos la función de distancia en lugar de restar tuplas
      minimo = distancia_2D(sublista[0], sublista[1])
      ele1 = sublista[0]
      ele2 = sublista[1]
      for i in range(len(sublista)):
        for j in range(i+1,len(sublista)):
          if( minimo > distancia_2D(sublista[i],sublista[j])):
            minimo = distancia_2D(sublista[i],sublista[j])
            ele1 = sublista[i]
            ele2 = sublista[j]
      return ele1, ele2, minimo


    # Encontramos el punto medio
    mitad = len(sublista) // 2
    punto_medio = sublista[mitad]
    ele1_izq, ele2_izq, min_izq = resolver(sublista[:mitad])
    ele1_der, ele2_der, min_der = resolver(sublista[mitad:])

    # Elegimos el mínimo entre la mitad derecha y la mitad izquierda
    if min_izq < min_der:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_izq, ele2_izq, min_izq
    else:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_der, ele2_der, min_der

    # Al ordenar por el eje x, en el punto medio pueden tener valores iguales en la coordenada x
    # pero en la coordenada y hay valores, es decir una franja
    # Buscamos puntos que estén a una distancia menor a minimo_elegido del centro
    franja = [p for p in sublista if abs(p[0] - punto_medio[0]) < minimo_elegido]

    # Ordenamos la franja por la coordenada Y
    franja.sort(key=lambda p: p[1])

    for i in range(len(franja)):
      for j in range(i + 1, len(franja)):
        # Si la diferencia de altura ya es mayor que minimo_elegido, no sigamos buscando
        if (franja[j][1] - franja[i][1]) >= minimo_elegido:
          break
        # Calculamos la distancia real
        distancia_franja = distancia_2D(franja[i], franja[j])
        if distancia_franja < minimo_elegido:
          minimo_elegido = distancia_franja
          ele1_elegido, ele2_elegido = franja[i], franja[j]

    return ele1_elegido, ele2_elegido, minimo_elegido

  resultado = resolver(lista_ordenada)
  return resultado[0], resultado[1]

```

```python
def cercanos_3D_divide_y_venceras(lista):

  if len(lista) < 2:
    return 'No hay un mínimo de dos puntos'

  # Primero ordenamos la lista sobre el eje x
  lista_ordenada = sorted(lista, key=lambda p: p[0])

  def distancia_3D(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

  def resolver(sublista):
    # Iremos reduciendo la lista hasta tener grupos de 2 o 3
    if len(sublista) <= 3:
      # ERROR CORREGIDO: Usamos la función de distancia en lugar de restar tuplas
      minimo = distancia_3D(sublista[0], sublista[1])
      ele1 = sublista[0]
      ele2 = sublista[1]
      for i in range(len(sublista)):
        for j in range(i+1,len(sublista)):
          if( minimo > distancia_3D(sublista[i],sublista[j])):
            minimo = distancia_3D(sublista[i],sublista[j])
            ele1 = sublista[i]
            ele2 = sublista[j]
      return ele1, ele2, minimo

    # Encontramos el punto medio
    mitad = len(sublista) // 2
    punto_medio = sublista[mitad]
    ele1_izq, ele2_izq, min_izq = resolver(sublista[:mitad])
    ele1_der, ele2_der, min_der = resolver(sublista[mitad:])

    # Elegimos el mínimo entre la mitad derecha y la mitad izquierda
    if min_izq < min_der:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_izq, ele2_izq, min_izq
    else:
      ele1_elegido, ele2_elegido , minimo_elegido = ele1_der, ele2_der, min_der

    # Al ordenar por el eje x, en el punto medio pueden tener valores iguales en la coordenada x
    # pero en la coordenada y hay valores, es decir una franja
    # Buscamos puntos que estén a una distancia menor a minimo_elegido del centro
    franja = [p for p in sublista if abs(p[0] - punto_medio[0]) < minimo_elegido]

    # Ordenamos la franja por la coordenada Y
    franja.sort(key=lambda p: p[1])

    for i in range(len(franja)):
      for j in range(i + 1, len(franja)):
        # Si la diferencia de altura ya es mayor que minimo_elegido, no sigamos buscando
        if (franja[j][1] - franja[i][1]) >= minimo_elegido:
          break
        # Calculamos la distancia real
        distancia_franja = distancia_3D(franja[i], franja[j]) #Clave para no tener que filtrar por la coordenada z
        if distancia_franja < minimo_elegido:
          minimo_elegido = distancia_franja
          ele1_elegido, ele2_elegido = franja[i], franja[j]

    return ele1_elegido, ele2_elegido, minimo_elegido

  resultado = resolver(lista_ordenada)
  return resultado[0], resultado[1]
```
