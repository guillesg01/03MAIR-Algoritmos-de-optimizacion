- ¿Como represento el espacio de soluciones?

El espacio de soluciones se representa mediante un vector de 10 números enteros $$V = [v_1, v_2, \dots, v_{10}]$$

donde:

-Índice $i$: Representa el partido de la jornada según el orden de la tabla de datos.

-Valor $v_i$: Es un entero $\in \{0, 1, \dots, 9\}$ que identifica el horario asignado.

- ¿Cual es la función objetivo?

La función objetivo es:
$$F = \sum_{i=1}^{10} Base_i ⋅ Ponderación_j ⋅ ( 1- Correción_k) $$

Donde:
- $\text{Base}_i$: Audiencia según las categorías de los equipos (A, B o C) en horario base:
$$\text{Base}_i = \begin{cases} 2.00 & \text{si A-A} \\ 1.30 & \text{si B-A (o A-B)} \\ 1.00 & \text{si A-C (o C-A)} \\ 0.90 & \text{si B-B} \\ 0.75 & \text{si B-C (o C-B)} \\ 0.47 & \text{si C-C} \end{cases}$$

- $\text{Ponderación}_j$: Coeficiente del horario $j$ asignado:
$$\text{Ponderación}_j = \begin{cases} 1.00 & \text{si } j \in \{\text{S20, D20}\} \\ 0.85 & \text{si } j = \text{D18} \\ 0.80 & \text{si } j = \text{S18} \\ 0.75 & \text{si } j = \text{D16} \\ 0.70 & \text{si } j = \text{S16} \\ 0.55 & \text{si } j = \text{S12} \\ 0.45 & \text{si } j = \text{D12} \\ 0.40 & \text{si } j \in \{\text{V20, L20}\} \end{cases}$$

- $\text{Corrección}_k$: Reducción por el número de coincidencias $k$ en el mismo horario:
$$\text{Corrección}_k = \{0, 0.25, 0.45, 0.60, 0.70, 0.75, 0.78, 0.80, 0.80\} \text{ para } k \in \{0, 1, \dots, 8\}$$

- ¿Como implemento las restricciones?

Hay dos restricciones:
1) Unicidad de Asignación: Para cada partido un único horario:

$$ \sum_{j=1}^{10} x_{i,j} = 1 \quad \forall i \in \{1, \dots, 10\} $$

donde:
$$x_{i,j} = \begin{cases} 1 & \text{si el partido } i \text{ se asigna al horario } j \\ 0 & \text{en caso contrario} \end{cases}$$


2) Obligatoriedad de Jornada: Debe haber un partido mínimo en los horarios de viernes y lunes: Sea $H =\{V20,L20\}$:
  $$\sum_{i=1}^{10} \mathbb{1}(v_i = horario_j) \geq 1, \quad \forall horario_j \in H$$

#Análisis
- ¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones


Dado que cada uno de los 10 partidos puede ser asignado a cualquiera de los 10 horarios disponibles (permitiendo coincidencias), el número total de combinaciones posibles es:$$ 10^{10}$$

Pero el hecho que debe haber al menos un partido el lunes y otro el viernes (la tabla de coincidencias da la opción de que haya 8, es decir, 9 partidos en el mismo horario) modifica el espacio de soluciones a:

$$\text{Válidas} = \text{Total} - \text{(ningún viernes} + \text{ningún lunes} -\text{ningún viernes ni lunes)}$$

$$ \text{Válidas} = 10^{10} - 9^{10} - 9^{10} + 8^{10} = 4.100.173.022 $$

Luego el espacio de soluciones es de 4.100.173.022 soluciones.

Una búsqueda exhaustiva que evaluara todas las combinaciones tendría coste:

$$O(H^N)$$

Con $H = 10$ horarios y $N = 10$ partidos, lo que crece exponencialmente con el número de partidos.

# Diseño
- ¿Qué técnica utilizo? ¿Por qué?

Se utilizará un Algoritmo Genético debido al gran tamaño del espacio de soluciones y a la no linealidad de la función objetivo, ya que mover un partido de horario modificará la audiencia de todos los demás que coincidan en esa misma franja, creando dependencias entre variables que los métodos lineales no pueden capturar.

Gracias a los operadores de cruce y mutación, el algoritmo explorará el espacio de soluciones de forma amplia sin quedarse atrapado en óptimos locales, combinando calendarios exitosos y probando cambios aleatorios para alcanzar configuraciones que maximicen la audiencia total. Por último, las restricciones obligatorias de viernes y lunes se gestionarán directamente en la generación de la población inicial y en la función de fitness, garantizando que la evolución se centre únicamente en soluciones válidas.



```
# Importaciones
import random
```

```
# Datos conocidos
partidos = [
    {"p": "Celta-R.Madrid", "cat": "BA", "base": 1.3},
    {"p": "Valencia-R.Sociedad", "cat": "BA", "base": 1.3},
    {"p": "Mallorca-Eibar", "cat": "CC", "base": 0.47},
    {"p": "Athletic-Barcelona", "cat": "BA", "base": 1.3},
    {"p": "Leganés-Osasuna", "cat": "CC", "base": 0.47},
    {"p": "Villarreal-Granada", "cat": "BC", "base": 0.75},
    {"p": "Alavés-Levante", "cat": "BB", "base": 0.9},
    {"p": "Espanyol-Sevilla", "cat": "BB", "base": 0.9},
    {"p": "Betis-Valladolid", "cat": "BC", "base": 0.75},
    {"p": "Atlético-Getafe", "cat": "BB", "base": 0.9}
]
horarios =[
    {"h": 0, "d": "V20", "pond": 0.4}, {"h": 1, "d": "S12", "pond": 0.55},
    {"h": 2, "d": "S16", "pond": 0.7}, {"h": 3, "d": "S18", "pond": 0.8},
    {"h": 4, "d": "S20", "pond": 1.0}, {"h": 5, "d": "D12", "pond": 0.45},
    {"h": 6, "d": "D16", "pond": 0.75}, {"h": 7, "d": "D18", "pond": 0.85},
    {"h": 8, "d": "D20", "pond": 1.0}, {"h": 9, "d": "L20", "pond": 0.4}
]
coincidencias = {0: 0.0, 1: 0.25, 2: 0.45, 3: 0.60, 4: 0.70, 5: 0.75, 6: 0.78, 7: 0.80, 8: 0.80}
```

```
# Definición de la función fitness
def fitness(lista):
    conteo_horarios = {}
    for h_id in lista:
        conteo_horarios[h_id] = conteo_horarios.get(h_id, 0) + 1

    # 2)Restricción: Un partido el viernes (h=0) y otro el lunes (h=9)
    if conteo_horarios.get(0, 0) == 0 or conteo_horarios.get(9, 0) == 0:
        return 0.0

    audiencia_total = 0
    for i, h_id in enumerate(lista):
        base = partidos[i]["base"] # Base según categoría
        pond = horarios[h_id]["pond"] # Ponderación del horario

        # El número de coincidencias es (total en ese horario - 1)
        num_coincidencias = conteo_horarios[h_id] - 1
        reduccion = coincidencias.get(num_coincidencias, 0.80) # % de reducción

        # Función a maximizar
        audiencia_total += (base * pond) * (1 - reduccion)

    return audiencia_total
```

```
def crear_individuo_valido():
    ind = [random.randint(1, 8) for _ in range(10)]  # Solo sábado/domingo
    ind[random.randint(0, 9)] = 0   # Forzar al menos un viernes
    ind[random.randint(0, 9)] = 9   # Forzar al menos un lunes
    return ind
```

```
# Función de cruce
def cruzar(p1, p2):
    punto = random.randint(1, 8)
    return p1[:punto] + p2[punto:]
```

```
# Función de mutación
def mutar(lista, tasa=0.1):
    if random.random() < tasa:
        idx = random.randint(0, 9)
        nuevo_horario = random.randint(0, 9)

        # Comprobar si el cambio rompe alguna restricción
        viernes_restantes = lista.count(0) - (1 if lista[idx] == 0 else 0)
        lunes_restantes = lista.count(9) - (1 if lista[idx] == 9 else 0)

        # Solo mutar si no deja sin viernes ni sin lunes
        if viernes_restantes >= 1 and lunes_restantes >= 1:
            lista[idx] = nuevo_horario
        # Si no, forzar que la mutación sea segura
        else:
            posiciones_seguras = [
                i for i in range(10)
                if not (lista[i] == 0 and lista.count(0) == 1)
                and not (lista[i] == 9 and lista.count(9) == 1)
            ]
            if posiciones_seguras:
                lista[random.choice(posiciones_seguras)] = random.randint(0, 9)

    return lista
```

```
def resolver_jornada(tam_pob=100, n_gen=1000, paciencia=100):

    # Crear población inicial con individuos válidos
    poblacion = [crear_individuo_valido() for _ in range(tam_pob)]

    sin_mejora    = 0
    mejor_fitness = 0.0

    for gen in range(n_gen):

        # Ordenar población de mejor a peor según audiencia
        poblacion = sorted(poblacion, key=fitness, reverse=True)
        fitness_actual = fitness(poblacion[0])

        # Actualizar contador de generaciones sin mejora
        if fitness_actual > mejor_fitness:
            mejor_fitness = fitness_actual
            sin_mejora = 0
        else:
            sin_mejora += 1

        # Parar si llevamos 'paciencia' generaciones sin mejorar
        if sin_mejora >= paciencia:
            break

        # Elitismo: los 10 mejores pasan directamente a la siguiente generación
        nueva_gen = poblacion[:10]

        while len(nueva_gen) < tam_pob:
            # Selección por torneo: elegir el mejor de 3 candidatos aleatorios
            padre1 = max(random.sample(poblacion[:50], 3), key=fitness)
            padre2 = max(random.sample(poblacion[:50], 3), key=fitness)

            # Cruzar los padres y mutar el hijo resultante
            hijo = mutar(cruzar(padre1, padre2))
            nueva_gen.append(hijo)

        poblacion = nueva_gen

    # Devolver el mejor individuo y su audiencia total
    return poblacion[0], fitness(poblacion[0])
```

```
mejor_plan, max_audiencia = resolver_jornada()

print(f"\n{'Partido':<25} {'Horario':<8} {'Base':>6} {'Pond':>6} {'Audiencia':>10}")
print("-" * 60)

conteo_horarios = {}
for h_id in mejor_plan:
    conteo_horarios[h_id] = conteo_horarios.get(h_id, 0) + 1

for i, h_id in enumerate(mejor_plan):
    base    = partidos[i]["base"]
    pond    = horarios[h_id]["pond"]
    num_c   = conteo_horarios[h_id] - 1
    reduccion = coincidencias.get(num_c, 0.80)
    aud     = base * pond * (1 - reduccion)
    print(f"{partidos[i]['p']:<25} {horarios[h_id]['d']:<8} {base:>6.2f} {pond:>6.2f} {aud:>9.2f}M")

print("-" * 60)
print(f"{'TOTAL':<25} {'':<8} {'':>6} {'':>6} {max_audiencia:>9.2f}M")
```

El algoritmo genético implementado logra una audiencia total de 6,81 millones de espectadores. La distribución resultante muestra una clara estrategia de maximización: los tres partidos de mayor interés (Celta-R.Madrid, Athletic-Barcelona y Valencia-R.Sociedad, todos categoría B-A) acaparan el 54% de la audiencia total con 3,70M, al ser asignados a las mejores franjas disponibles (D18, S20 y D20 con coeficientes de 0,85 y 1,0). En el extremo opuesto, los dos partidos de menor audiencia potencial quedan relegados a los horarios obligatorios de viernes y lunes: Betis-Valladolid (V20, 0,30M) y Mallorca-Eibar (L20, 0,19M), minimizando así el impacto de esas franjas de bajo rendimiento.

Nota sobre la variabilidad de resultados:

El algoritmo genético implementado es de naturaleza estocástica, es decir, utiliza decisiones aleatorias en varias de sus etapas: la generación de la población inicial, la selección por torneo, el punto de cruce y la mutación. Esto implica que cada ejecución puede producir un resultado ligeramente diferente, aunque siempre cercano al óptimo.
